```python
import requests
import datetime as dt
import pytz
```

```python

coordinates = [41.6500201, -72.7342163]

time_str = "00:00"
moon_date_list =[]
for i in range(10):
    current_date = (dt.date.today() + dt.timedelta(days=i)).strftime("%Y-%m-%d")
    celestial_URL = f"https://aa.usno.navy.mil/api/celnav?date={current_date}&tz=-9&time={time_str}&coords={coordinates[0]},{coordinates[1]}"
    celestial_data = requests.get(celestial_URL).json()
    moon_illum = celestial_data['properties']['moon_illum']
    moon_phase = celestial_data['properties']['moon_phase']
    moon_date_list.append({
        'date': current_date,
        'moon_illumination': moon_illum,
        'moon_phase': moon_phase
    })
    print(f"Data for {current_date}: Moon Illumination - {moon_illum}, Moon Phase - {moon_phase}")
```

```python
coordinates = [36.8508, -75.2859]
time_str = "23:00"
for i in range(30):
    current_date = (dt.date.today() + dt.timedelta(days=i)).strftime("%Y-%m-%d")
    celestial_URL = f"https://aa.usno.navy.mil/api/celnav?date={current_date}&time={time_str}&coords={coordinates[0]},{coordinates[1]}"
    celestial_data = requests.get(celestial_URL).json()

    if 'properties' in celestial_data:
        moon_illum = celestial_data['properties'].get('moon_illum')
        moon_phase = celestial_data['properties'].get('moon_phase')

        if moon_illum is not None and moon_phase is not None:
            print(f"Data for {current_date}: Moon Illumination - {moon_illum}, Moon Phase - {moon_phase}")
        else:
            new_time_str = "02:00"
            new_celestial_URL = f"https://aa.usno.navy.mil/api/celnav?date={current_date}&time={new_time_str}&coords={coordinates[0]},{coordinates[1]}"
            new_celestial_data = requests.get(new_celestial_URL).json()
            new_moon_illum = new_celestial_data['properties'].get('moon_illum')
            new_moon_phase = new_celestial_data['properties'].get('moon_phase')
            print(f"Data for {current_date} (updated time): Moon Illumination - {new_moon_illum}, Moon Phase - {new_moon_phase}")
    else:
        print(f"No moon data available for {current_date}")
```

```python
base_url = 'https://api.met.no/weatherapi/sunrise/2.0/.json'

current_date = dt.datetime.now()
end_date = current_date + dt.timedelta(days=10)

while current_date < end_date:
    date_str = current_date.strftime('%Y-%m-%d')

    params = {
        'lat': 59.911491,  # Replace with your latitude
        'lon': 10.757933,  # Replace with your longitude
        'date': date_str,
    }

    response = requests.get(base_url, params=params)

    if response.status_code == 200:
        data = response.json()
        moon_illumination = data['location']['time'][0]['moon_illumination']
        moon_phase = data['location']['time'][0]['moonphase']['@name']

        print(f"Date: {date_str}, Moon Illumination: {moon_illumination}, Moon Phase: {moon_phase}")
    else:
        print(f"Data retrieval failed for date: {date_str}")

    current_date += dt.timedelta(days=1)
```
